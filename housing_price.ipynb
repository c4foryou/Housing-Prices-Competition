## Outlier check

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (16,8))
sns.scatterplot(train['GrLivArea'], train['SalePrice'])
```

```python
# Deleting 'bad' Outliers in 'GarageArea' column

di = train[(train['GarageArea'] > 1220) & (train['SalePrice'] < 300000)].index
train = train.drop(index = di)
train[(train['GarageArea'] > 1220) & (train['SalePrice'] < 300000)].index
```

## Imputation for missing values

```python
all_data = pd.concat([train, test])
all_data2 = all_data.drop(columns=['Id','SalePrice', 'Alley', 'PoolQC', 'Fence', 'MiscFeature'])
all_data2['MSSubClass'] = all_data2['MSSubClass'].astype('str')

cat_columns = all_data2.columns[all_data2.dtypes == object]

all_data2[cat_columns] = all_data2[cat_columns].fillna(-1)


num_columns = all_data2.columns[all_data2.dtypes != object]

for n in num_columns:
    all_data2[n] = all_data2[n].fillna(all_data2[n].median())

all_data2.info()
```

## One-hot-encoding for categorical columns

```python
all_data3 = pd.get_dummies(all_data2)

all_data3 = all_data3.fillna(-1)
all_data3
```

## Scaling by using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
all_data4 = ss.fit_transform(all_data3)
all_data4 = pd.DataFrame(all_data4, columns = all_data3.columns)
all_data4
```

## Final datasets

```python
train2 = all_data4[:len(train)]
train2
```

```python
test2 = all_data4[len(train):]
test2
```

# Modeling (Ridge)

## Cross validation and Hyperparameter tuning for 'alpha'

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

rmse = {}
for i in [10,20,30,50,100,200,300,500,700,1000]:
    rid_cv = Ridge(alpha = i)
    cvs_ht = cross_val_score(rid_cv, train2, train['SalePrice'], n_jobs = -1, cv = 10, scoring = 'neg_mean_squared_error') # 회귀 에러에서는 neg를 붙어야됨. 그래야 베스트 모델 잘찾아줌
    rmse[i] = (cvs_ht.mean()*-1)**(1/2)

rmse
```

## Final model

```python
# using best alpha value for the final model

rid = Ridge(alpha = 200)

# taking log for y value when fitting the model

rid.fit(train2, np.log(train['SalePrice']))
```

## Prediction and Train RMSE

```python
from sklearn.metrics import mean_squared_error

result = rid.predict(test2)
train_rmse = (mean_squared_error(train['SalePrice'], np.exp(rid.predict(train2))))**(1/2)
train_rmse
```

```python
sub = pd.read_csv('sample_submission.csv')
sub['SalePrice'] = np.exp(result)
sub.to_csv('sub.csv', index = 0)
```
